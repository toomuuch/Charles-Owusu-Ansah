**Question 5: Create a code block that performs the following operations: (10 points)**

   1. Import the `bitcoin_cash_price.csv` file into a Pandas dataframe called `bitcoin`.
   
   2. Import the `dash_price.csv` file into a Pandas dataframe called `dash`.
   
   3. Merge the `bitcoin` and `dash` dataframes using the `Date` column as the identifier column. Name this new merged dataframe `crypto`. 
   
   4. Create a new `Delta_Volume` column in the `crypto` dataframe that calculates the largest **absolute** difference between the bitcoin `Volume` versus dash `Volume`.
   
     * **Hint:** Use `abs()` to calculate absolute value.
   
   
   5. Find the row in the `crypto` dataframe that contains the maximum `Delta_Volume` value. Print the row out. 

````python
#QUESTION 5
# Answer 1
file1 = pd.read_csv("../Unsolved/bitcoin_cash_price.csv")
bitcoin = pd.DataFrame(file1)
bitcoin.head()
````

````python
#QUESTION 5
# Answer 2
file2 = pd.read_csv("../Unsolved/dash_price.csv")
dash = pd.DataFrame(file2)
dash.head()
````

````python
#QUESTION 5
# Answer 3
# using merge function to merge the bitcoin and dash dataframes using date column as identifier 
crypto = pd.merge(bitcoin,dash,on="Date") 
crypto.head()
````

````python
#QUESTION 5
# Answer 4
diff = crypto["Volume_x"] - crypto["Volume_y"] # this calculates the difference in the two volumes and assigned to diff
diff2 = abs(diff)  #this returns the absolute values of the diff(the difference in the volumes)
crypto["Delta_Volume"]=diff2 # this creates new column in crypto called Delta_volume and put the abs values in the column
crypto.head()
````

````python
#QUESTION 5
# Answer 5
diff3=max(diff2) #calculates the max value of the abs values in Delta_Volume column
df=crypto.where(crypto["Delta_Volume"]==diff3).dropna() # using where funtion to filter the crypto for max value 
df
````
