# Hypothesis 
<ol> 
<li> A day with no Precipatation and Higher Temperature affect bike riding by increasing the number of bikes crossing brooklyn bridge.


# Null Hypothesis
<ol> 
<li> Precipitation and Temperature does not affect bikers in anyway, no precipatation and higher temperature does not contribute to the number of bikes cross brooklyn bridge.

```python
# Dependencies 
import pandas as pd
import plotly.express as plt
import os
from scipy.stats import ttest_1samp
import numpy as np
import scipy.stats as sts
import plotly.express as px
import statsmodels.api as sm
from sklearn import datasets
```

```python
bikecounts_file = os.path.join('../','Resources', 'bikecounts.csv')
bikecounts_df = pd.read_csv(bikecounts_file)
bikecounts_df['Day'] = pd.to_datetime(bikecounts_df['Day'])
bikecounts_df.head()
```

```python
#checking data type of the columns of the dataframe
bikecounts_df.info()
```

# meteorology
<lo>
<li> In meteorology, a trace T denotes an amount of precipitation, 
<li> such as rain or snow, that is greater than zero, but is too small to be measured by standard units or methods of measurement.
<li> though T values are very small it can still have societal impact using 0.0005 so it can be used among non zero precipitation data

```python
#definig a function which will take away (S) and replace T with 0.0005 when applied
#convert all datatype from object to float

def units_cov(val):
    new_val =val.replace('(S)','').replace('T','0.0005')
    return float(new_val)
    
```

```python
#apply the function on the precipation column to make it float64 
bikecounts_df['Precipitation']=bikecounts_df['Precipitation'].apply(units_cov)
```

```python
#viewing the DF to see the efect of the fuction
bikecounts_df.head()
```

```python
#rechecking the data types after applying the function
bikecounts_df.info()
```

```python
#Runing Linear regression model
# Select the independent X (temperature & Precipitation) terms, and the  dependent y (number of bikes) term. 
X = bikecounts_df[['High Temp (°F)','Low Temp (°F)','Precipitation']]
y = bikecounts_df['Brooklyn Bridge']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
```

```python
# Print the linear regression results
print(results.summary())
```

# Result 1 
<ol> 
<li> The result opbtained in the linear regression shows the P values, which indicates that High temperatures does affect the number of bikes crossing the bridge, also the precipation affects the number of bikes too.
<li> The p values support the hypothesis and rejects the null hypothesis,
<li> however i cant tell if the day with no precipation increases the bikes count or not, precipitation  pvalues only shows all days.
